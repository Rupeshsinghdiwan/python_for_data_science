```python
x = np.arange(0, 10*np.pi, 0.1) ## 0.1 is diff betwwen value
y= np.sin(x)
plt.plot(x,y ,c="r")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("sin graph")
```

```python
x = np.arange(0,5*np.pi,0.2)
y_sin =np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,"r--")
plt.title("sine grap")
plt.subplot(2,1,2)
plt.plot(x,y_cos,"g--")
plt.title("cosin graph")
```

```python
x = [2,8,10] 
y = [11,16,9]  
x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y ,color='r') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
################  Histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
############# Box Plot using Matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

```python
# Pie Chart
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
